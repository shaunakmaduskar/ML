```python
test_data=pd.read_csv('/kaggle/input/marketing-strategy-personalised-offer/test_data.csv')
train_data=pd.read_csv('/kaggle/input/marketing-strategy-personalised-offer/train_data.csv')
sample=pd.read_csv('/kaggle/input/marketing-strategy-personalised-offer/sample.csv')
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
simpleimputer = SimpleImputer(strategy='most_frequent')
d={}
c=train_data.columns
for i in train_data.columns :
  d[i] = sum(train_data[i].value_counts())
for i in d.keys():
  if d[i] <=10000 :
    print(i)
    train_data[i] = labelencoder.fit_transform(train_data[i])
  
train_data = simpleimputer.fit_transform(train_data)
train_data = pd.DataFrame(train_data,columns=c)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
for i in test_data.columns :
  #print(test_data[i].value_counts())
  test_data[i] = labelencoder.fit_transform(test_data[i])
for i in train_data.columns :
  #print(train_data[i].value_counts())
  train_data[i] = labelencoder.fit_transform(train_data[i])
```

```python
y = train_data['Offer Accepted']
X = train_data[train_data.columns.drop('Offer Accepted')]
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state =0)
```

```python
X_T = test_data
X_T
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
cl = HistGradientBoostingClassifier()
cl.fit(X_train,y_train)
cl.score(X_test,y_test)
y_pred = cl.predict(X_T)
```

```python
submission=pd.DataFrame(columns=['id','Offer Accepted'])

submission['id']=[i for i in range(len(y_pred))]


submission['Offer Accepted']=['Yes' if y_pred[i] == 1 else 'No' for i in range(len(y_pred))]

# Export it as an csv
submission.to_csv('submission.csv',index=False)
```
